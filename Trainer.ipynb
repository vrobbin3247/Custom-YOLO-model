```python
from ultralytics import YOLO
import os

# Load YOLOv8 model
model = YOLO('yolov8n.pt')

# Train the model and save outputs to Google Drive
results = model.train(
    data='mydata.yml',
    imgsz=640,
    epochs=10,
    batch=32,
    name='yolov8n_custom'
)
```

```python
model_path = 'runs/yolov8m_custom/weights/best.pt'
model_saved = YOLO(model_path)
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_path = 'sample.jpg'

results = model_saved(image_path)


result = results[0]

annotated_image = result.plot()

plt.imshow(cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

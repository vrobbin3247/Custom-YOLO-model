```python
import os
import random
import matplotlib.pyplot as plt
from collections import Counter
from PIL import Image

# Define paths and classes
images_path = 'bdd100/images-mini'
labels_path = 'bdd100/labels-mini-txt'
class_names = ['car', 'bus', 'person', 'bike', 'truck', 'motor', 'train', 'rider', 'traffic sign', 'traffic light']

# Count images
image_files = [f for f in os.listdir(images_path) if f.endswith(('.jpg', '.png'))]
print(f"Total images: {len(image_files)}")

# Count class distribution
class_counts = Counter()

for label_file in os.listdir(labels_path):
    if label_file.endswith('.txt'):
        with open(os.path.join(labels_path, label_file), 'r') as f:
            for line in f:
                class_id = int(line.split()[0])
                class_counts[class_names[class_id]] += 1

print("Class distribution:")
for class_name, count in class_counts.items():
    print(f"{class_name}: {count}")

# Show random sample images (up to 5)
sample_images = random.sample(image_files, min(5, len(image_files)))
plt.figure(figsize=(15, 5))
for i, img_file in enumerate(sample_images):
    img_path = os.path.join(images_path, img_file)
    img = Image.open(img_path)
    plt.subplot(1, len(sample_images), i + 1)
    plt.imshow(img)
    plt.axis('off')
    plt.title(img_file)

plt.show()
```

```python
import os
import random
import matplotlib.pyplot as plt
from collections import Counter
from PIL import Image

# Define paths and classes
images_path = 'bdd100/images-mini'
labels_path = 'bdd100/labels-mini-txt'
class_names = ['car', 'bus', 'person', 'bike', 'truck', 'motor', 'train', 'rider', 'traffic sign', 'traffic light']

# Count images
image_files = [f for f in os.listdir(images_path) if f.endswith(('.jpg', '.png'))]
print(f"Total images: {len(image_files)}")

# Count class distribution
class_counts = Counter()

for label_file in os.listdir(labels_path):
    if label_file.endswith('.txt'):
        with open(os.path.join(labels_path, label_file), 'r') as f:
            for line in f:
                class_id = int(line.split()[0])
                class_counts[class_names[class_id]] += 1

print("Class distribution:")
for class_name, count in class_counts.items():
    print(f"{class_name}: {count}")

# Plot class distribution as a bar graph with counts on bars
plt.figure(figsize=(10, 6))
bar_plot = plt.bar(class_counts.keys(), class_counts.values(), color='skyblue')
plt.xlabel('Classes')
plt.ylabel('Count')
plt.title('Class Distribution in BDD100K Mini Dataset')
plt.xticks(rotation=45)

# Add counts on top of bars
for bar in bar_plot:
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.1, bar.get_height() - 0.1, str(bar.get_height()), ha='center', va='bottom', fontsize=12, color='black', fontweight='bold')

plt.show()

# Show random sample images (up to 5)
sample_images = random.sample(image_files, min(5, len(image_files)))
plt.figure(figsize=(15, 5))
for i, img_file in enumerate(sample_images):
    img_path = os.path.join(images_path, img_file)
    img = Image.open(img_path)
    plt.subplot(1, len(sample_images), i + 1)
    plt.imshow(img)
    plt.axis('off')
    plt.title(img_file)

plt.show()
```
